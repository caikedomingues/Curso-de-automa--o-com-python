```python

import os

from dotenv import load_dotenv

import requests
```

```python
load_dotenv()

CHAVE_API = os.getenv('CHAVE_API')
```

Link da documentação do alpha vantage: https://www.alphavantage.co/documentation/

Pegando cotações semanais - JSON

```python


url = f'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol=IBM&apikey={CHAVE_API}'
r = requests.get(url)
data = r.json()

print(data)
```

```python
print(data["Meta Data"])
```

Pegando cotações semanais - CSV

```python
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol=IBM&apikey={CHAVE_API}&datatype=csv'
r = requests.get(url)

print(r.text)
```

```python
import pandas as pd

from io import StringIO

tabela = pd.read_csv(StringIO(r.text))

display(tabela)
```

Cotações

```python
acoes = ['ITUB4', 'ABEV3', 'BBAS3']
```

```python
compilada = pd.DataFrame()
for acao in acoes:
    
    url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={acao}.SAO&apikey={CHAVE_API}&datatype=csv'
    r = requests.get(url)

    cotacoes = pd.read_csv(StringIO(r.text))
    
    compilada = pd.concat([compilada, cotacoes])

display(compilada)
```

Encontrar Ativos

```python
pesquisa = input("Informe o nome da empresa que você esta procurando")

url = f'https://www.alphavantage.co/query?function=SYMBOL_SEARCH&keywords={pesquisa}&apikey={CHAVE_API}&datatype=csv'
r = requests.get(url)

ativo = pd.read_csv(StringIO(r.text))

display(ativo)
```

Informações

```python
import pprint

url = f'https://www.alphavantage.co/query?function=EARNINGS&symbol=AMZN&apikey={CHAVE_API}&datatype=json'

r = requests.get(url)

data = r.json()

pprint.pprint(data)

resultado = pd.DataFrame(data['annualEarnings'])

display(resultado)
```
