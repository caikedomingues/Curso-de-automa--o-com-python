Lendo varias abas em arquivos excel

Import da biblioteca pandas

```python
# Biblioteca que possibilita a leitura e a manipulação de dados presentes em um arquivo
# excel.
import pandas as pd
```

Leitura padrão do pandas

```python
# Na leitura padrão do pandas, usamos a função read_excel para ler as tabelas presentes na
# planilha, porém, o read_excel lê apenas a primeira tabela que ele encontra na primeira aba
# do excel(primeira tabela da esquerda para a direita). Por conta disso, temos que passar outros
# argumentos na função para garantir que o read_excel encontre outras abas e tabelas. 
planilha = pd.read_excel('arquivo.xlsx')

# Ira mostrar as tabelas encontradas.
display(planilha)
```

Lendo uma planilha("aba") especifica, com indice de aba:

```python
# Para ler os dados de uma aba especifica, vamos passar como valor do
# argumento sheet_name o nome da aba que queremos ler os dados.
planilha = pd.read_excel('arquivo.xlsx', sheet_name='Planilha2')

display(planilha)
```

Lendo uma planilha ("aba") especifica, com indice da aba

```python
# Caso você não saiba o nome da aba que você precisa, basta você passar como 
# argumento o indice da aba para o argumento sheet_name. Observação: A xontagem
# no python começa a partir do 0.
planilha = pd.read_excel('arquivo.xlsx', sheet_name=3)

display(planilha)
```

Selecionando as colunas pelo indice

```python
# Além de especificarmos a aba, também podemos especificar as colunas que
# queremos acessar. Para isso, usamos o argumento usecols que recebe uma
# lista de valores numéricos que representam os indices da colunas
planilha = pd.read_excel('arquivo.xlsx', sheet_name='Planilha3', usecols=[5, 6])

display(planilha)
```

Selecionando as colunas pelo "nome da coluna"

```python
# Também podemos especificar as colunas através do nome delas (que também
# podem ser representadas em intervalos, como fizemos nesse caso). Para
# isso basta passar como valor do argumento usecols, o nome ou o intervalo
# que você quer acessar.
planilha = pd.read_excel('arquivo.xlsx', sheet_name='Planilha3', usecols="F:G")

display(planilha)
```

Buscando todas as informações de um arquivo

```python
# Para coletarmos todos os dados de todas as abas, precisamos passar
# como valor do argumento sheet_name o valor None. Dessa forma, não
# especificaremos colunas e a função read_excel retornara um dicionário
# com todos os dados da planilha, onde as abas serão as chaves, e o valor
# das tabelas serão os valores do dicionário, por exemplo, a aba "Planilha1"
# será a chave de acesso para os valores presentes na aba.
planilha = pd.read_excel('arquivo.xlsx', sheet_name=None)

display(planilha)
```

Buscando informações de uma tabela fora da célula "A1"

Maneira padrão que não funciona

```python
# Como vimos anteriormente, o read_excel por padrão lê apenas a primeira tabela
# da esquerda para a direita, para resolver isso, aprendemos a utiliza o argumento
# sheet_name e usecols, porém eles não são o suficiente quando as linhas estão distantes
# da célula A1, ou seja, distantes do "radar de leitura do read_excel". Quando isso acontece,
# o read_excel retorna valores nulos nas linhas que ele não identificou.
planilha = pd.read_excel('arquivo.xlsx', sheet_name='Planilha4')

display(planilha)
```

Como resolver esse problema

```python
# Para resolver esse problema, uma das alternativas é fazer o read_excel 
# ignorar as linhas que não contém os dados da aba (tabela de valores).
# Dessa forma ele começara a ler o arquivo a partir da existência dos 
# dados, para isso usaremos o argumento skiprows que recebe um valor inteiro
# que irá indicar a quantidade de linhas que o read_excel irá pular
# na hora da leitura
planilha = pd.read_excel('arquivo.xlsx', sheet_name='Planilha4', usecols="G:H",skiprows=7)

display(planilha)
```

Especificando a quantidade de linhas que o read_excel deve ler

```python
# Para especificar a quantidade de linhas que a função deve ler, devemos passar
# como argumento o nrows que recebe um valor inteiro que indica a quantidade
# de linhas que devem ser lidas
planilha = pd.read_excel('arquivo.xlsx', sheet_name='Planilha5', nrows=11)

display(planilha)
```

Buscando apenas as informações desejadas (quando usamos o None no sheet_name)

```python
# Quando o read_excel retorna um dicionário (usando o none no argumento sheet 
# name), podemos acessar os valores de uma aba especifica usando o nome
# da chave que contém os valores (nome da aba que queremos acessar)
planilha = pd.read_excel('arquivo.xlsx', sheet_name=None)

display(planilha['Planilha2'])
```
