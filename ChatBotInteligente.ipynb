Import das Bibliotecas Necessárias para o Exercicio

```python
# Classe da biblioteca que possui que possibilita a criação
# de chatbots
from chatterbot import ChatBot

# Import da classe ListTrainer do módulo trainers da biblioteca chatterbot
# Que tem como objetivo criar uma lista de valores que serão utilizados para
# treinar as respostas do chatbot.
from chatterbot.trainers import ListTrainer

# Import da classe POR do módulo languages (módulo de linguagens e idiomas
# do chatterbot) da biblioteca chatterbot que tem como objetivo configurar
# o idioma do chatbot em português.
from chatterbot.languages import POR
```

Criando o ChatBot

```python
# Instancia do chatbot que criaremos no exercicio, ela recebe como argumento
# em seu construtor:
# BotCaike: nome do chatbot que construiremos
# tagger_language = POR: Irá conter o idioma do chatbot 
chatbot = ChatBot("BotCaike", tagger_language=POR)

# Irá conter a lista de valores que treinarão a resposta do chatbot
conversa = [
    
    "Coe",
    
    "E ai, tranquilo?",
    
    "Tranquilo",
    
    "Qual a boa de hoje?",
    
    "Estou estudando automação em Python",
    
    "Caracas, ai sim",
    
    "Maneiro né",
    
    "Irado"
]

# Instância da classe ListTrainer que irá treinar o chatbot. Ele
# recebe como argumento do construtor, o chatbot que será treinado
# pela classe.
trainer = ListTrainer(chatbot)

# Função da classe ListTrainer que irá treinar o chatbot usando como
# argumento a lista de conversa. O treinamento funciona da seguinte 
# maneira, o chatbot sempre segue a seguinte sequência: mensagem e resposta,
# ou seja, ele considera uma linha como "mensagem do usuário" e a próxima como
# "resposta do chatbot".
trainer.train(conversa)
```

Conversando com o ChatBot

```python
# Vamos iniciar agora o envio de mensagens ao chatbot 

# Ira criar um loop infinito que possibilitara o envio de inumeras mensagens
# ao chatbot.
while True:
    
    # Ira coletar as mensagens do usuário. O upper ira converter todas
    # as letras de minusculo para maiusculo. Dessa maneira fica mais
    # fácil de validar as entradas e verificar se o chatbot deve
    # ou não continar respondendo ao usuário.
    mensagem = input("Mande uma mensagem para o chatbot:").upper()
    
    if mensagem == "PARAR":
        
        # Se a mensagem do usuário for "PARAR" vamos dar um break que
        # encerrará a execução do loop e de coneversas com o chat    
        print("Conversa Encerrada")
        break
    
    else:
        
        # Caso contrário, o usuário poderá conversar infinitamente com o
        # chat
        
        # get_response: Função da classe ChatBot que irá coletar a mensagem
        # do usuário e devolver uma resposta (baseada na lista de treinamento
        # utilizada no train).
        resposta = chatbot.get_response(mensagem)
        
        # Ira imprimir a resposta do chatbot
        print(resposta)

# Observação: O sistema irá armazenar um banco de dados sqlite as respostas
# (do sistema) e mensagens do usuário com o objetivo de continuar aprimorando
# o seu aprendizado
```
