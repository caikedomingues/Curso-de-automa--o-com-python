API IBGE: É uma ferramenta gratuita que permite que desenvolvedores e 
pesquisadores acessem dados oficiais do Brasil diretamente em suas 
aplicações, sem precisar baixar planilhas manualmente. Um dos pontos
positivos da API é o retorno dos dados em formato JSON que pode ser
facilmente manipulado pelo python usando a biblioteca requests

Principais serviços:

    -> API de Localidades: Permite listar todos os estados, municipios,
    regiões e distritos do Brasil. É muito usada para criar aqueles
    menus de seleção de cidade/estado em sites.

    -> API de Nomes (Censos): Permite consultar a frequência de um nome 
    por década, descobrir em qual estado um nome é mais comum e ver o ran
    king de nomes.

    -> API de agregados(SIDRA): Acesso a tabelas complexas do censo, indicadores de inflação (IPCA), desemprego (PNAD) e PIB.

    -> API de Malhas Geográficas: Retorna arquivos (SVG ou TopoJSON)
    para desenhar mapas do Brasil ou de cidades especificas.

Consultando valores na API 

```python
# Biblioteca que possui funções para envio de requisições ao servidor.
import requests as r

# Ira conter o link de requisição que é gerado no site da api do ibge. O
# link contém as informações que filtramos no site.
link = "https://servicodados.ibge.gov.br/api/v3/agregados/4112/periodos/2006/variaveis/2586?localidades=N1[all]"

# Função que tem como onbjetivo realizar uma requisição get (consulta
# de dados) a API. 
requisicao = r.get(link)

# Ira retornar as informações no formato JSON que contém uma lista de
# dicionários com inumeras informações sobre o tema especificado no link
informacoes = requisicao.json()

# Ira printar o json com as informações
print(informacoes)
```

```python
# Import da biblioteca ppprint que tem como objetivo imprimir os dados de forma
# organizada
import pprint

pprint.pprint(informacoes)
```

Acessando o primeiro dicionário da lista

```python
# Para especificar o tipo de dicionário que queremos acessar, devemos
# utilizar o indice do dicionário (posição na lista) 
pprint.pprint(informacoes[0])
```

Acessando valores especificos

```python
# Para acessar valores especificos devemos indicar para o comando 
# o indice do dicionário que conytém o valor e o rótulo do valor procurado
item_busca = informacoes[0]['variavel']

resultados = informacoes[0]['resultados'][0]['series']

pprint.pprint(item_busca)

pprint.pprint(resultados)
```
