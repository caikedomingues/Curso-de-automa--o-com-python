```python
# O nome vem de Simple Mail Transfer Protocol (smtp)

# Para que serve: Ela serve para criar a conexão de rede entre o computador
# e o servidor do Google (ou qualquer outro servidor de e-mail).

# função tecnica:Ela abre o canal de comunicação, faz login usando a nossa 
# senha de app e empurra os dados para a nuvem para que eles cheguem ao
# destinatário.
import smtplib

# Enquanto a smtplib cuida do transporte, esta biblioteca cuida do conteúdo.

# para que serve: Ela usada para montar o e-mail de forma que os servidores
# consigam ler.

# Função Técnica: Ela organiza os metadados do e-mail. Sem ela, você enviaria
# apenas um texto solto; com ela, você define tecnicamente o que é Assunto
# (subject), quem é o remetente (From) e quem é o destinatario (To)
import email.message

# Biblioteca que tem como objetivo conectar o python ao nosso sistema de arquivo.
# Vamos usar ele para acessar as variáveis de ambientes.
import os

# Função que irá carregar a variável de ambiente no nosso código
from dotenv import load_dotenv  
```

```python
# Carregando a variável de ambiente no código
load_dotenv()

# Ira coletar o emaui da variável de ambiente EMAIL.
EMAIL = os.getenv('EMAIL')

# Ira coletar a senha de app da variável de ambiente SENHA
SENHA = os.getenv('SENHA')
```

Criando a função que envia emails

```python
# A função não precisará de argumentos
def enviar_email():
    
    # Ira conter a mensagem/conteúdo do email
    corpo_email = "<p>Ola Caike</p> <p>Email recebido</p>"
    
    # Ira armazenar o objeto Message que contém um dicionário de
    # valores que devem conter em um email    
    mensagem = email.message.Message()
    
    # Ira definir o assunto do email
    mensagem['Subject'] = "Assunto"
    
    # Ira definir o remetente
    mensagem['From'] = EMAIL
    
    # Ira definir o destinatário.
    mensagem['To'] = EMAIL
    
    # Senha de app que o smtp usará para logar no email e enviar a mensagem
    password = SENHA
    # Irá criar o cabeçalho da mensagem (irá definir onde cada conteúdo deve ficar)
    mensagem.add_header('Content-Type', 'text/html')
    
    # Ira carregar na mensagem o conteudo do email
    mensagem.set_payload(corpo_email)
    
    # Ira gerenciar a porta de segurança do email
    smt = smtplib.SMTP('smtp.gmail.com: 587')
    
    # Irá criptografar a senha antes de enviar a senha para o servidor do google
    smt.starttls()
    
    # Ira realizar login no gmail usando como argumento
    # o email e a senha de app.
    smt.login(mensagem['From'], password)
    
    # Ira enviar o email usando como argumento o o remetente e o destinatario.
    # as_string: Indica que é uma mensagem de texto
    # encode: Codifica a mensagem no padrão utf-8
    smt.sendmail(mensagem['From'], mensagem['To'], mensagem.as_string().encode('utf-8'))
    
    print('Email enviado')
```

```python
enviar_email()
```
