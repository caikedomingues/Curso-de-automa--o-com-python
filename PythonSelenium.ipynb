                            Automação em Python

-> A automação com Python é o uso da linguagem para criar scripts (pequenos
programas) que executam tarefas repetitivas, manuais ou previsiveis no computador, simulando a ação humana ou comunicando-se diretamente com sistemas.

-> Em vez de clicar, digitar e mover arquivos, o Python faz isso por você
com uma velocidade e precisão infinitamente maiores.

                            Pilares da Automação
1. Automação de Interface (RPA - Robotic Process Automation): Aqui o Python
"toma o controle" do seu mouse e teclado. Ele abre programas, clica em
botões e preenche formulários em softwares que não possuem uma forma fácil
de integração.

    -> Exemplo: Abrir um sistema de contabilidade antigo e cadastrar 500
    notas fiscais manualmente.

    -> Bibliotecas Principais: PyAutoGUI, Pywinauto

2. Automação Web: É focada no navegador. O python abre o Chrome/Firefox, navega em sites, faz login, extrai dados (web scraping) ou preenche formulários online.

    -> Exemplo: Entrar no site de 10 concorrentes toda manhã, pegar os preços
    de um produto e salvar em uma planilha.

    -> Bibliotecas Principais: Selenium, Playwright, BeatifulSoup

3. Automação de Dados e Arquivos: É a manipulação de arquivos "por baixo
do capô", sem necessariamente abrir a interface visual. É extremamente rápida.

    -> Exemplo: Ler 100 arquivos Excel, filtrar apenas as vendas ácima de
    1.000,00 gerar um PDF de relatório para cada vendedor.

    -> Bibliotecas Principais: Pandas, Openpyxl, Os, Shutil.

                        Automação Web com Selenium

Selenium: É um conjunto de ferramentas focado na automação de navegadores
web. Ele permite que você escreva um codigo em python que "comanda" o navegador como se fosse um humano: clicando em botões, digitando em campos
de texto e lendo o que está escrito na página.

    -> O que ele faz: Envia comandos para o navegador.

    -> A "peça" central: Para o Selenium funcionar, ele precisa de um 
    driver (um tradutor). O chrome tem o ChromeDriver, o Firefox tem o
    GeckoDriver, e assim por diante. Sem esse driver o Python não consegue
    conversar com o navegador.

WebDriver Manager: Este é o herói anônimo que resolve um dos maiores problemas
de quem começa na automação.

    -> O Problema: Antigamente, você precisava baixar o chromedriver.exe manualmente, conferir se a versão do seu Chrome era a mesma do driver
    e colocar o caminho do arquivo no código. Se o seu Chrome atualizasse 
    sozinho (o que ele faz sempre), sua automação quebrava por que as versões
    deixavam de ser compativeis.

    -> A Solução (WebDriver Manager): É uma biblioteca que faz todo esse trabalho sujo automaticamente. Quando você inicia seu código ele:

        1. Checa a versão do seu navegador atual

        2. Verifica se o driver correto já está instalado

        3. Se não estiver (ou se estiver desatualizado), ele baixa a versão
        correta na hora e a configura para o python usar.

```python
# Biblioteca que contém os comandos para clicar, digitar e navegar.
from selenium import webdriver

# Será o nosso assistente de manutenção. Ele irá garantir que nós 
# tenhamos o driver correto para a versão do nosso chrome.
from webdriver_manager.chrome import ChromeDriverManager

# É uma classe que o Selenium usa para configuarar como o driver do 
# navegador deve ser iniciado no sistema operacional (Instalação
# e atualizaçaõ do drive).
from selenium.webdriver.chrome.service import Service

# Instância da classe Service que irá receber como argumento do
# construtor o chromeDriverManager que irá verificar a versão do
# seu navegador e instalar o drive compativel com ele
servico = Service(ChromeDriverManager().install())

# Instância do objeto que será responsável por manipular o nosso navegador.
# Ele recebe como argumento do construtor, o driver instalado (armazenado
# na variável servico) que conecta o python com o navegador
navegador = webdriver.Chrome(service=servico)
```

Acessando os elementos de um Site

```python
# Ira coletar a pesquisa que o usuário quer realizar no youtube
pesquisa_youtube = input("Escreva o que você quer pesquisar no youtube")

# Ira verificar se o valor informado no campo de pesquisa é válido para
# ser enviado ao youtube.
status_pesquisa = True

if pesquisa_youtube == "":
    
    # Se o campo de pesquisa for vázio, vamos imprimir mensagem e 
    # atribuir ao status de pesquisa o valor False.    
    print("A pesquisa não pode ser em branco")
    
    status_pesquisa = False

if status_pesquisa == True:
    
    # Se o status da pesquisa for true, significa que a entrada informada é
    # válida e podemos envia-la ao navegador.
    
    # find_element: Método da classe navegador(webdriverchrome) que consegue
    # acessar os elementos html de uma página com o objetivo de acessar o
    # elemento do site. A função recebe como argumento;
    
    # xpath: Como o find_element irá procurar os elementos

    # '//*[@id="center"]/yt-searchbox/div[1]/form/input: Xpath do elemento que 
    # iremos acessar (Nesse caso será a barra de pesquisa do YouTube)

    # send_keys: Função do elemento encontrado que possibilita enviar entradas
    # ao navegador. A função recebe como argumento o input que coleta entradas
    # de usuário. 
    
    # Conceito de xpath: Permite que você encontre qualquer "casa" (elemento
    # ) usando coordenadas especificas, mesmo que essa casa não tenha um 
    # número de id ou uma classe.
    navegador.find_element('xpath', '//*[@id="center"]/yt-searchbox/div[1]/form/input').send_keys(pesquisa_youtube)
    
    # Ira conter o mesmo comando do código anterior, porém, dessa vez utilizaremos o click que será responsavel por clicar no botão de 
    # pesquisa do  youtube. Observação: O argumento do find_element será
    # o xpath do botão de pesquisa. 
    navegador.find_element('xpath', '//*[@id="center"]/yt-searchbox/button').click()
    
```
