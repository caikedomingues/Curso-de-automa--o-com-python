Conhecendo algumas funções da biblioteca os

```python
# Biblioteca do python que tem como objetivo acessar o sistema de arquivos
# do sistema operacional
import os

# listdir: Função que tem como objetivo listar os arquivos e pastas presentes
# em um diretório. Por padrão, se a função não recebe um caminho como argumento,
# ela exibe os arquivos da pasta que o código python esta salvo (Diretório
# de Trabalho Atual).
arquivos = os.listdir()

print(arquivos)
```

Verificando a pasta que estamos trabalhando

```python
# Função que coleta o caminho da pasta que estamos trabalhando
caminho = os.getcwd()

print("Caminho da pasta que estamos trabalhando: ",caminho)
```

```python
# Função da biblioteca os que permite que o python renomeie os arquivos
# do diretório acessado. A função recebe como argumento o nome atual
# do arquivo/diretório e o nome que você quer colocar no item. 
os.rename('vendas - 1.xlsx', 'Vendas - 1.xlsx')
```

```python
# A função rename também pode mover os arquivos de um lugar para outro.
# Para isso devemos passar como argumento o caminho atual do diretório
# e o caminho de destino do arquivo que queremos mover
os.rename('Vendas - 1.xlsx', 'Vendas/Vendas - 1.xlsx')
```

```python
# É uma extensão da biblioteca os, focada em operações de alto nivel sobre
# arquivos e coleções de arquivos. Enquanto a os cuida de tarefas mais
# básicas do sistema, a shutil facilita ações complexas como copiar, mover
# e remover diretórios inteiros
import shutil

# Função da biblioteca shutil que tem como objetivo copiar arquivos.
# A função recebe como argumento o caminho original e o caminho de
# destino da cópia.
shutil.copy('Vendas - 1.xlsx', 'Vendas/Vendas - 1.xlsx')
```

Listando os arquivos da pasta do projeto

```python
lista_arquivos = os.listdir()

arquivos
```

```python
# For que irá percorrer a lista de arquivos e pastas do diretório de 
# trabalho atual.
for arquivo in lista_arquivos:
    
    # Como podemos observar, essa lista contém dados do tipo texto,
    # logo, é possivel verificar palavras chaves na construção do
    # código
    
    # Vamos verificar se a extensão do arquivo é do tipo xlsx (arquivos
    # excel).
    if 'xlsx' in arquivo:
        
        # Se essa condição for verdadeira, vamos verificar se o nome
        # dos arquivos contém a palavra "Compra".
        if "Compras" in arquivo:
            
            # Se essa condição for verdadeira, vamos mover os arquivos
            # para a pasta de compras.
            os.rename(arquivo, f'Compras/{arquivo}')
        
        elif "Vendas" in arquivo:
            # Se essa condição for verdadeira, vamos mover os arquivos para
            # a pasta de "vendas"    
            os.rename(arquivo, f'Vendas/{arquivo}')
            
        
```
