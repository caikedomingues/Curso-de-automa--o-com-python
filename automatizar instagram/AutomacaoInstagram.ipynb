```python

# Biblioteca que possui os drivers de conexão do python com os navegadores.
# Vamos utilizar para acessar o driver de conexão do google chrome
from selenium import webdriver

# Import da biblioteca que possibilita que o python acesse o nosso sistema
# de arquivos. Vamos utilizar ela para acessar a variável de ambiente que possui
# a senha do nosso instagram
import os

# Import da função que irá carregar as variaveis de ambientes no projeto 
from dotenv import load_dotenv

# Import da classe By do selenium que possui os seletores que utilizaremos
# para selecionar os elementos que queremos controlar no navegador.
from selenium.webdriver.common.by import By

# Biblioteca de manipulação de tempo de execução, vamos usar ela para
# dar intervalos na execução do código com o objetivo de garantir que
# a página ja´tenha sido completamente carregada. Dessa forma, evitaremos
# erros de elementos não encontrados
import time

# Esta classe serve para fazer o selenium ficar vigiando o navegador por
# até x segundos. Ele não trava o código por todo o tempo; ele fica
# verificando a página repetidamente em intervalos curtissimos. Assim
# que o que você quer acontece, o código segue imediatamente, economizando
# tempo.
from selenium.webdriver.support.ui import WebDriverWait
 
# Geralmente importada como EC para facilitar a escrita, ela contém uma
# lista de "regra" ou "eventos" que você esta esperando acontecer. Alguns
# exemplos comuns:

# presence_of_element_located: Espera o elemento aparecer no código HTML

# element_to_be_clickable: Espera o elemento não só aparecer, mas esta 
# visivel e pronto para receber um clique (sem nada na frente dele).

# visibility_of_element_located: Espera o elemento ficar visivel na tela. 
from selenium.webdriver.support import expected_conditions as EC
```

Acessando as variáveis de ambiente

```python
# Ira carregar a variável no código
load_dotenv()

# Ira coletar o valor da variável de ambiente
SENHA_INSTAGRAM = os.getenv('SENHA_INSTAGRAM')
```

Criando o Navegador

```python
# Irá instanciar a classe Chrome que possui os métodos que manipulam os elementos
# de um navegador chrome
navegador = webdriver.Chrome()
```

Acessando o site do instagram

```python
# Ira usar o método get para acessar o instagram no navegador chrome
navegador.get("https://www.instagram.com/")
```

Marcando a opção agora não

```python
# Ira instanciar a classe WebDriverWait que analisa se a página ja foi
# devidamente carregada no navegador. O construtor da página irá receber
# como argumento o objeto (navegador) e um valor numerico (10) que representa o tempo máximo de espera (em segundos). O "intervalo de verificação" (o tempo que ele espera entre uma olhada e outra na página) é um valor padrão de 0,5 segundos, a menos que você o altere.
aguardar = WebDriverWait(navegador, 10)

# Ira inspecionar o bloco de código com o objetivo de capturar possiveis erros.
try:
    
    # until: metodo da classe webdriverwait que especifica o que a classe
    # deve procurar
    
    # EC.presence_of_element_located: Ira especificar que a execução deve
    # aguardar o elemento aparacer na tela para prosseguir.
    # //div: Procure uma divisão de bloco.
    # [@role='button']: Que tenha a função de botão
    # and contains(text(), 'Agora'): E que tenha a palavra "Agora" escrita dentro
    botao_agora = aguardar.until(
        EC.presence_of_element_located((By.XPATH, "//div[@role='button' and contains(text(), 'Agora')]"))
    )
    

    # Injeta o comando diretamente no motor do navegador (JavaScript). Ele diz
    # "não me importa se tem algo na frente ou se o mouse não esta lá, apenas
    # execute a função de clique desse elemento agora". É quase impossivel para
    # o site bloquear esse tipo de clique 
    navegador.execute_script("arguments[0].click();", botao_agora)
    print("Botão 'Agora não' clicado com sucesso!")

except Exception as e:
    
    # Ira capturar e tratar o erro moatrando para o usuário o tipo de erro encontrado
    print(f"Não foi possível clicar no botão: {e}")
```
