Schedule: É uma biblioteca do Python extremamente popular por sua simplicidade. Ela foi desenhada para ser "amigável aos humanos", 
permitindo o agendamento de tarefas recorrentes usando uma sintaxe que 
parece uma frase em inglês. Imagine que você criou um robô e quer que
ele limpe o banco de dados todo domingo a meia-noite ou envie um relatório
de vendas em cada 10 minutos. O schedule é a ferramenta que controla esse
relógio.

Como Funciona: O funcionamento se baseia em 3 pilares:

    -> Função: O que deve ser feito (ex:enviar_email())

    -> Intervalo: Quando deve ser feito (ex: a cada 5 segundos, toda segunda-feira).

    -> Loop de Verificação: Um ciclo que fica pergutando ao sistema: "ja deu o horário?"

A sintaxe human-freindly: A grande vantagem dessa biblioteca é a leitura do
código. Veja como os comandos são intuitivos:

    -> A cada 10 minutos: schedule.every(10).minutes.do(função)
    -> Toda hora: schedule.every().hour.do(função)
    -> Diariamente as 10:30: schedule.every().day.at("10:30").do(função)
    -> A cada 5 a 10 minutos(aleatório): schedule.every(5).to(10).minutes.do(função).
