```python
# Permite escolher qual a impressora iremos utilizar. Essa biblioteca faz parte 
# de um pacote chamado pywin32. Ela serve exclusivamente para o python
# controlar o subsistema de impressão do windows. O que ela faz:
# Lista todas as impressoras instaladas (fisicas e virtuais).
# Identifica qual é a impressora padrão atual.
# Altera a impressora padrão via código.
# Monitora filas de impressão (ver se o papel acabou ou se há erro).
import win32print

# Ira permitir enviar comandos ao windows. É uma biblioteca de "baixo nivel"
# que permite ao python enviar comandos diretamente para a API principal
# do windows (a Win32 API).
import win32api

# Permite que o python acesse o gerenciador de arquivos do sistema operacional
# (no nosso caso, o sistema de arquivos do windows).
import os
```

```python
# Escolher a impressora

# EnumPrinters: Função que tem como objetivo enumerar (listar) as impressoras
# Ela recebe como argumento o numero 2 que indica para a função buscar
# apenas as impressoras locais. Se usassemos outros numeros, ele poderia
# buscar impressoras da rede ou servidores de impressão.
lista_impressora = win32print.EnumPrinters(2)

# For que irá percorrer a tupla que contém as tuplas que possuem as informações
# das impressoras.
for impressora in lista_impressora:
    
    # Ira imprimir a tupla de informações de impressoras.
    print(impressora)

```

```python
# ira selecionar a tupla que contém a impressora que escolhemos.
impressora_selecionada = lista_impressora[2] 

# Ira acessar o nome da impressora na tupla escolhida.
nome_impressora = impressora_selecionada[2]

# Função que torna a impressora escolhida como padrão do sistema operacional.
# Quando executamos essa função qualquer programa que o usuário abrir depois
# (Word, Excel, o próprio navegador) vai sugerir essa impressora automaticamente
# como a primeira opção de impressão.
win32print.SetDefaultPrinter(nome_impressora)
```

```python
# Ira conter o caminho da pasta que contém os arquivos
caminho = 'arquivos'

# listdir: Irá listar os arquivos presentes na pasta especificada.
# A função recebe como argumento o caminho da pasta que comtém os arquivos.
lista_arquivos = os.listdir(caminho)

print(lista_arquivos)
```

```python
# For que irá percorrer a lista de arquivos do diretório especificado
# na variável de lista de arquivos. Esse for irá executar a impressão
# de cada arquivo da pasta.
for arquivo in lista_arquivos:
    
    # Função que tem como objetivo imprimir os arquivos:
    
    # 0: É o identificador da janela pai. Usar o 0 indica que não há
    # uma janela especifica associada a esse comando.
    
    # print: Comando que a função manda ao sistema operacional (comando
    # de impressão).
    
    # arquivo: Nome do arquivo que será impresso.
    
    # None: Parâmetros extras. Para impressões simples, deixamos vázio.
    
    # caminho: Caminho da pasta que contém os arquivos
    
    # 0: Modo de exibição da janela. O valor 0 (conhecido como SE_HIDE)
    # serve para que o processo ocorra em segundo plano, sem abrir o
    # programa (como o WOrd ou o Adobe Reader) na cara do usuário.
    win32api.ShellExecute(0, "print", arquivo, None, caminho, 0)
```
